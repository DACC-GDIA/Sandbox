```python
def  writeB(df):
    try:
        conn = sqlite3.connect(dbname)
        cur = conn.cursor()
        for (index, record) in df.iterrows():
            NewUsed = record[2]
            Make = record[3]
            Model = record[4]
            Disposal = record[6]
            Fuel = record[8]
            Finance = record[9]

            cur.execute('''INSERT OR IGNORE INTO NewUsedB (Indicator) VALUES ( ? )''', ( NewUsed, ) )
            cur.execute('SELECT NewUsedKey FROM NewUsedB WHERE Indicator = ? ', (NewUsed, ))
            NewUsed_id = cur.fetchone()[0]  

            cur.execute('''INSERT OR IGNORE INTO MakeB (Make) VALUES ( ? )''', ( Make, ) )
            cur.execute('SELECT MakeKey FROM MakeB WHERE Make = ? ', (Make, ))
            Make_id = cur.fetchone()[0]

            cur.execute('''INSERT OR IGNORE INTO ModelB (Model) VALUES ( ? )''', ( Model, ) )
            cur.execute('SELECT ModelKey FROM ModelB WHERE Model = ? ', (Model, ))
            Model_id = cur.fetchone()[0]

            cur.execute('''INSERT OR IGNORE INTO DisposalB (Disposal) VALUES ( ? )''', ( Disposal, ) )
            cur.execute('SELECT DisposalKey FROM DisposalB WHERE Disposal = ? ', (Disposal, ))
            Disposal_id = cur.fetchone()[0]

            cur.execute('''INSERT OR IGNORE INTO FuelTypeB (FuelType) VALUES ( ? )''', ( Fuel, ) )
            cur.execute('SELECT FuelTypeKey FROM FuelTypeB WHERE FuelType = ? ', (Fuel, ))
            Fuel_id = cur.fetchone()[0]

            cur.execute('''INSERT OR IGNORE INTO FinancingB (Financing) VALUES ( ? )''', ( Finance, ) )
            cur.execute('SELECT FinancingKey FROM FinancingB WHERE Financing = ? ', (Finance, ))
            Finance_id = cur.fetchone()[0]

            cur.execute('''INSERT OR IGNORE INTO B (CustID, PurchaseDate, NewUsed, Make,
            Model, ModelYear, Disposal, DisposalDate, FuelType, Financing) VALUES ( ?, ?, ?, ?, ?, ?, ?, ?, ?, ? )''',
                        (record[0], record[1], NewUsed_id, Make_id, Model_id, record[5],
                         Disposal_id, record[7], Fuel_id, Finance_id, ) )
    
    except:
        traceback.print_exc()
        print(index, record[0], record[1], NewUsed_id, Make_id, Model_id, record[5],
                         Disposal_id, record[7], Fuel_id, Finance_id)
        cur.close()
        conn.close()
        
    conn.commit()
    cur.close()
    conn.close()
```

```python
def CleanA(df):
    criteria = df.dtypes == "object"
    df[criteria.index[criteria]] = df[criteria.index[criteria]].fillna(value = "None")
    df[criteria.index[~criteria]] = df[criteria.index[~criteria]].fillna(value = 0)
    return df
```

```python
def CleanB(df):
    df[0] = df[0].fillna(value = 0)
    df[1] = df[1].fillna(value = 0)
    df[2] = df[2].fillna(value = "None")
    df[3] = df[3].fillna(value = "None")
    df[4] = df[4].fillna(value = "None")
    df[5] = df[5].fillna(value = 0)
    df[6] = df[6].fillna(value = "None")
    df[7] = df[7].fillna(value = 0)
    df[8] = df[8].fillna(value = "None")
    df[9] = df[9].fillna(value = "None")
    return df
```
