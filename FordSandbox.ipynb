This is a filename agnostic way to import the data. Just replace the path with the location of your data.

```python
path = r"C:\Users\bjr21\Documents\GitHub\Sandbox\Data"
files = os.listdir(path)
```

You can modify this below to import only a certain file (e.g. use files[5]) or to iterate through all files to either append or paginate the reading

```python
for file in files:
    if "A0" in file:
        df = pd.read_csv(os.path.join(path,file), header=None)
```

```python
df.info()
```

```python
df = df.dropna(axis=1, how='all')
```

```python
criteria = df.dtypes == "object"
```

```python
df[criteria.index[~criteria]] = df[criteria.index[~criteria]].fillna(value = 0)
```
